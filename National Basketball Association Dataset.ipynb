# Data Modeling

### For this section, I first use the groupby function by calling it gk and using it to group NBA teams together, further separating the dataset. I then define what gk is, in order to actually get out the printed dataset I want (taht lists the grouped Team names). Next, I chose a NBA team to focus on and delve deeper in using the learned codes in order to gather the data I want from only the Miami Heat team. Then, I use the info and describe function to go even deeper into the data for Miami Heat and to help myself analyze the data for that team even better.

```python
# use this function to group dataset into NBA teams
gk = df.groupby('Team')

# use type to identify gk
type (gk)

# use this code to list grouped Teams alphabetically
gk.first()
```

```python
# this function is used to list data for only the Miami Heat 
gk.get_group('Miami Heat')
```

```python
# this code is used to look at number of columns and their names for Miami Heat, as well as finding any missing data from the columns
gk.get_group('Miami Heat').info()
```

```python
# this function is used to find the overview of numerical numbers in max, min, avrg, etc. pertaining to data for Miami Heat
gk.get_group('Miami Heat').describe()
```
